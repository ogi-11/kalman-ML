```python
import pandas as pd
import pandas_datareader.data as web
import datetime

# 1. Define the start and end dates for your project
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime(2024, 12, 31)

# 2. Download the VIX data directly from the FRED database
# The ticker for VIX on FRED is 'VIXCLS'
try:
    ir_df = web.DataReader(['DGS10', 'DGS3MO'], 'fred', start_date, end_date)
    
    # 3. Rename the column for clarity
    ir_df.rename(columns={'DGS10': '10-Year Treasury Constant Maturity Rate', 'DGS3MO': '3-Month Treasury Constant Maturity Rate'}, inplace=True)
    
    print("Successfully downloaded IR data from FRED.")
    print(ir_df.head(30))

except Exception as e:
    print(f"Failed to download data from FRED. Error: {e}")
    print("You may need to run 'pip install pandas-datareader' first.")

ir_df.to_csv('/Users/ogizelenovic/Downloads/interestrates.csv', index=True)
```
